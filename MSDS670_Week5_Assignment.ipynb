```python
#Import Libraries
import os

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

%matplotlib inline
```

```python
#setting up out dataframe
df = pd.read_csv('C:/Users/kyleb/Downloads/iris.csv')
```

### Bar Chart
We are going to start by creating a boxplot of the features of our dataset.

```python
df.boxplot(['Sepal Length', 'Sepal Width ', 'Petal Length', 'Petal Width'], color = 'b')
plt.title("Features Boxplot")
plt.xlabel('Feature')
plt.ylabel('Length (cm)')
plt.grid(False)
```

We find that there a couple of outliers in the sepal length but they are not super extreme so we are going to leave them.

## Plot 2

```python
ax = sns.histplot(data=df, x = "Petal Length", hue = 'Species')
ax.set_xlabel('Petal Length (cm)')
ax.set_title('Petal Lengths by Species')
```

We see that the overwhelming majority of our short petals belong to the Setosa species while the Virginica have most of the longer petals.

## Plot 3

For our final plot we are going to try and see how length ratios are related to width ratios.

```python
df['length_ratio'] = df['Petal Length']/df['Sepal Length']
df['width_ratio'] = df['Petal Width']/df['Sepal Width ']
```

```python
ax = sns.scatterplot(data=df, x='length_ratio', y='width_ratio', hue='Species')
ax.set_title('Ratios of Lengths to Widths')
ax.set_xlabel('Length Ratio')
ax.set_ylabel('Width Ratio')
```
